```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
n1=imputer.fit_transform(dataset)
```

```python
dataset1=pd.DataFrame(n1,columns=dataset.columns)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
n2=std.fit_transform(dataset1)
```

```python
dataset2=pd.DataFrame(n2,columns=dataset1.columns)
```

```python
import sklearn.utils
dataset2=sklearn.utils.shuffle(dataset2)
dataset2.reset_index(drop = True)
```

```python
dataset2.head(11)
```

```python
dataset2.describe()
```

```python
x1=dataset2.iloc[:,:-1].values
y1=dataset2.iloc[:,-1].values
```

```python
x1
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x1,y1,test_size=0.2,random_state=0)
```

## Model  


## 1. Linear regression

```python
from sklearn.linear_model import LinearRegression
regressor1=LinearRegression()
regressor1.fit(xtrain,ytrain)
regressor1.predict(xtest)
```

```python
regressor1.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressor1,xtest,ytest,scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
print(scores.std())
```

## 2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor2=DecisionTreeRegressor()
regressor2.fit(xtrain,ytrain)
regressor2.predict(xtest)
```

```python
regressor2.score(xtest,ytest)
```

```python
scores = cross_val_score(regressor2,xtest,ytest,scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
scores.std()
```

## 3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor3=RandomForestRegressor()
regressor3.fit(xtrain,ytrain)
regressor3.predict(xtest)
```

```python
regressor3.score(xtest,ytest)
```

```python
scores = cross_val_score(regressor3,xtest,ytest,scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```
